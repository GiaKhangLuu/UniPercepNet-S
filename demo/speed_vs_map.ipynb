```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data
data = {
    'Method': ['Mask RCNN-R101', 'CenterMask', 'CenterMask-Lite', 'YOLACT-550-R50', 'YOLACT-550-R101', 'YOLACT-700', 'YOLOF Mask-R50', 'YOLOF Mask-R50-SE'],
    'FPS': ['6.4', '6.9', '9.1', '28.49', '20.13', '13.49', '12.66', '9.1'],
    'Mask mAP': ['37.9', '38.3', '32.9', '28.2', '29.8', '31.2', '30.9', '33.2']
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Convert FPS and Mask mAP to numeric values
df['FPS'] = pd.to_numeric(df['FPS'])
df['Mask mAP'] = pd.to_numeric(df['Mask mAP'])

# Sort the DataFrame by FPS
df = df.sort_values(by='FPS')

# Marker and Color settings for each method
markers = {'Mask RCNN-R101': 'o', 'CenterMask': '*', 'CenterMask-Lite': '*', 'YOLACT-550-R50': 'v', 'YOLACT-550-R101': 'v', 'YOLACT-700': 'v', 'YOLOF Mask-R50': 's', 'YOLOF Mask-R50-SE': 's'}
colors = {'Mask RCNN-R101': 'green', 'CenterMask': 'blue', 'CenterMask-Lite': 'blue',  'YOLACT-550-R50': 'orange', 'YOLACT-550-R101': 'orange', 'YOLACT-700': 'orange', 'YOLOF Mask-R50': 'red', 'YOLOF Mask-R50-SE': 'red'}

# Plotting
plt.figure(figsize=(8, 6))

# Plot each method with corresponding marker and color
for method in df['Method'].unique():
    subset = df[df['Method'] == method]
    plt.scatter(subset['FPS'], subset['Mask mAP'], label=method, s=100, marker=markers[method], color=colors[method])

# Adding the 'Real-time' shaded area
#plt.axvspan(30, plt.xlim()[1], color='lightblue', alpha=0.3)
#plt.text(32, 26, 'Real-time', color='lightblue', fontsize=16, fontweight='bold')

# Customizing the plot
plt.xlabel('FPS')
plt.ylabel('Mask mAP')
plt.legend(title='')

# Adding dotted grid lines
plt.grid(True, which='both', linestyle=':', linewidth=0.5)

# Display the plot
plt.show()
```
