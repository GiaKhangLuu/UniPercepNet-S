```python
from ultralytics import YOLO
```

```python
model = YOLO("yolo11s-seg.pt")

results = model.train(
    data='./datasets/bdd100k/labels_yolo/data.yaml',
    epochs=36,
    imgsz=(1333, 800),
    batch=2,
    device=0,
    workers=4,
    name='yolo11s_bdd100k',
    pretrained=False,
    optimizer='AdamW',
    lr0=1e-4,
    patience=70,
    degrees=8.0,
    translate=0.15,
    scale=0.2,
    shear=1.0,
    fliplr=0.5,
    mosaic=0.0,
    mixup=0.0,
    copy_paste=0.0)
```
